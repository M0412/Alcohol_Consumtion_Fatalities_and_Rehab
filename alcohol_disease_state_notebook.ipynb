```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from splinter import Browser
import matplotlib.pyplot as plt
```

```python
df =pd.read_html( "https://www.cdc.gov/mmwr/volumes/69/wr/mm6930a1.htm")
table1_df = df[0] 
```

```python
table1_df.columns = ['cause', 'total_alcohol_related_death','male_alcohol_death','female_alcohol_death','total_years_of_potential_life_loss', 'male_years_of_potential_life_loss', 'female_years_of_potential_life_loss']
table1_df= table1_df.iloc[1:]
table1_df
```

```python
table2_df = df[1]
table2_df.columns = ['State', 'alcohol_related_deaths', 'age_adjusted_alcohol_related_deaths', 'years_of_potential_life_lost', 'years_of_life_lost_per_1000', 'alcohol_related_years_of_life_lost']
#taking out US total because it skewed the data by a lot in bar graph
table2_df=table2_df.iloc[1:] 
table2_df
```

```python
states = table2_df['State']
alcohol_deaths = table2_df['alcohol_related_deaths']


plt.figure(figsize=(10, 6))  
plt.bar(states, alcohol_deaths)
plt.xlabel('State')
plt.ylabel('Alcohol-Related Deaths')
plt.title('Alcohol-Related Deaths by State')
plt.xticks(rotation=90)  
plt.tight_layout()  


plt.show()
```
