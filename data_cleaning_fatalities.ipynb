```python
# Reading all the 2022 data from CA, FL, and TX
ca_df = pd.read_csv("Data/Excel:CSV/Quick_Statistics_Results_CBHSQ_Data_SAMHSA_California_2022.csv")
ca_df
```

```python
# Extract the alcohol column only
ca_alcohol_df = ca_df.iloc[:,[0,2]]
ca_alcohol_df = ca_alcohol_df.rename(columns={"State: CA":'State', 'Alcohol Only':'California'})
ca_alcohol_df
```

```python
ca_alcohol_df
```

```python
# dropping out the category label to keep it clean with indexes information
indexes_to_drop = [2, 7, 22, 30]
ca_alcohol_df = ca_alcohol_df.drop(indexes_to_drop, axis=0)
ca_alcohol_df
```

```python
# transposing the data to store as JSON later
ca_alcohol_df = ca_alcohol_df.transpose()
ca_alcohol_df
```

```python
# adjust the rows to headers of the dataframe
ca_alcohol_df.columns = ca_alcohol_df.iloc[0]
ca_alcohol_df = ca_alcohol_df.iloc[1: , :]
ca_alcohol_df
```
