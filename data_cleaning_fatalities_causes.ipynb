Deaths and Years of Potential Life Lost From Excessive Alcohol Use — United States, 2011–2015
[link](https://www.cdc.gov/mmwr/volumes/69/wr/mm6930a1.htm)


```python
df =pd.read_html( "https://www.cdc.gov/mmwr/volumes/69/wr/mm6930a1.htm")
table1_df = df[0] 
```

```python
table1_df.columns = ['cause', 'total_alcohol_related_death','male_alcohol_death','female_alcohol_death','total_years_of_potential_life_loss', 'male_years_of_potential_life_loss', 'female_years_of_potential_life_loss']
table1_df= table1_df.iloc[1:]
table1_df.head()
```

```python
table1_df[["male_alcohol_death","male_alcohol_death(%)"]] = table1_df["male_alcohol_death"].str.split(" ", expand=True)
table1_df["male_alcohol_death(%)"] = table1_df["male_alcohol_death(%)"].str.strip("()")

table1_df[["female_alcohol_death","female_alcohol_death(%)"]] = table1_df["female_alcohol_death"].str.split(" ", expand=True)
table1_df["female_alcohol_death(%)"] = table1_df["female_alcohol_death(%)"].str.strip("()")

table1_df[["male_years_of_potential_life_loss","male_years_of_potential_life_loss(%)"]] = table1_df["male_years_of_potential_life_loss"].str.split(" ", expand=True)
table1_df["male_years_of_potential_life_loss(%)"] = table1_df["male_years_of_potential_life_loss(%)"].str.strip("()")

table1_df[["female_years_of_potential_life_loss","female_years_of_potential_life_loss(%)"]] = table1_df["female_years_of_potential_life_loss"].str.split(" ", expand=True)
table1_df["female_years_of_potential_life_loss(%)"] = table1_df["female_years_of_potential_life_loss(%)"].str.strip("()")

table1_df.head()
```

```python
table1_df = table1_df.replace("—", 0)
final_df = table1_df.fillna(0)
final_df.head(10)
```

```python
final_df.dtypes
```

```python
final_df["male_alcohol_death"] = final_df["male_alcohol_death"].str.replace(",","")
final_df["female_alcohol_death"] = final_df["female_alcohol_death"].str.replace(",","")
final_df["male_years_of_potential_life_loss"] = final_df["male_years_of_potential_life_loss"].str.replace(",","")
final_df["female_years_of_potential_life_loss"] = final_df["female_years_of_potential_life_loss"].str.replace(",","")
```

```python
final_df = final_df.dropna()
final_df.head()
```

```python
final_df["male_alcohol_death"] = final_df["male_alcohol_death"].astype("int64")
final_df["female_alcohol_death"] = final_df["female_alcohol_death"].astype("int64")
final_df["male_years_of_potential_life_loss"] = final_df["male_years_of_potential_life_loss"].astype("int64")
final_df["female_years_of_potential_life_loss"] = final_df["female_years_of_potential_life_loss"].astype("int64")

final_df["male_alcohol_death(%)"]=final_df["male_alcohol_death(%)"].astype("float64")
final_df["female_alcohol_death(%)"]=final_df["female_alcohol_death(%)"].astype("float64")
final_df["male_years_of_potential_life_loss(%)"]=final_df["male_years_of_potential_life_loss(%)"].astype("float64")
final_df["female_years_of_potential_life_loss(%)"]=final_df["female_years_of_potential_life_loss(%)"].astype("float64")
```

```python
final_df.dtypes
```

```python
final_df.to_json("Data/JSON_Files/Alcohol_Related_Deaths_Per_Cause_2011-2015.json", orient="records")
```

Drunk Driving Statistics [link](https://www.bankrate.com/insurance/car/drunk-driving/)

```python
browser = Browser('chrome')
```

```python
url = "https://www.bankrate.com/insurance/car/drunk-driving/#stats-2022"
browser.visit(url)
```

```python
html = browser.html
soup = BeautifulSoup(html, 'html.parser')
```

```python
table_body_1 = soup.find("table", class_="Table Table--text Table--fixed Table--bordered Table--spacing1 table-inline")
table_body_1
```

```python
headings = table_body_1.find_all('th', class_="table-inline__header-cell")

heading_list = []

for heading in headings:
    name = heading.text
    heading_list.append(name)

heading_list
```

```python
rows = table_body_1.find_all("tr")

row_list = [
    [x.text for x in row.find_all("td")]
    for row in table_body_1.find_all("tr")
    if row.find_all("td")  # Only include rows with table data
]

row_list
```

```python
age_bracket_df = pd.DataFrame(row_list, columns=heading_list)
age_bracket_df
```

```python
age_bracket_df["% of alcohol-impaired drivers involved in fatal crashes"]=age_bracket_df["% of alcohol-impaired drivers involved in fatal crashes"].str.strip("%").astype("float64")
age_bracket_df.head()
```

```python
age_bracket_df.to_json("Data/JSON_Files/Alcohol_Fatalities_by_Age_Bracket_2022.json", orient="records", force_ascii=False, default_handler=str)
```

```python
table_body_2 = soup.find_all("table", class_="Table Table--text Table--bordered Table--spacing1 table-inline")
main = table_body_2[1]
```

```python
headings = main.find_all("th", class_="table-inline__header-cell")

headings_list = [heading.text for heading in headings]
headings_list
```

```python
rows = main.find_all("tr")

row_list = [
    [x.text for x in row.find_all("td")]
    for row in main.find_all("tr")
    if row.find_all("td")  # Only include rows with table data
]

row_list
```

```python
state_fatalities = pd.DataFrame(row_list, columns=headings_list)

state_fatalities.head()
```

```python
state_fatalities.dtypes
```

```python
state_fatalities["Total fatalities"]=state_fatalities["Total fatalities"].str.replace(",","").astype("int64")
state_fatalities["Total alcohol-related fatalities"]=state_fatalities["Total alcohol-related fatalities"].str.replace(",","").astype("int64")
state_fatalities["% of all fatalities involving alcohol"]=state_fatalities["% of all fatalities involving alcohol"].str.strip("%").astype("float64")
```

```python
state_fatalities.to_json("Data/JSON_Files/Alcohol_Fatalities_by_State_2022.json", orient="records")
```

FBI - Crime in the US in 2019 (Alcohol Charges) [link](https://ucr.fbi.gov/crime-in-the-u.s/2019/crime-in-the-u.s.-2019/topic-pages/tables/table-38)

```python
data = pd.read_csv("Data/Excel:CSV/Crime_by_Age_2019.csv", header=4)
data.head(10)
```

```python
data_alcohol_charges = data.loc[(data["Offense charged"]=="Driving under the influence")|
                                (data["Offense charged"]=="Liquor laws")|
                                (data["Offense charged"]=="Drunkenness")]

data_alcohol_charges
```

```python
new_columns = [col.replace('\n', ' ') for col in data_alcohol_charges.columns]
data_alcohol_charges.columns = new_columns
data_alcohol_charges = data_alcohol_charges.reset_index(drop=True)
data_alcohol_charges
```

```python
data_alcohol_charges.to_json("Data/JSON_Files/Alcohol_Charges_by_Age_2019.json", orient="records")
```
