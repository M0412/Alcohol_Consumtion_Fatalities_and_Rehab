```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from splinter import Browser
import matplotlib.pyplot as plt
import html5lib
```

```python
# Reading all the 2022 data from CA, FL, and TX
ca_df = pd.read_csv("samsha_data/Quick_Statistics_Results_CBHSQ_Data_SAMHSA_California_2022.csv")
ca_df
```

```python
# Extract the alcohol column only
ca_alcohol_df = ca_df.iloc[:,[0,2]]
ca_alcohol_df = ca_alcohol_df.rename(columns={"State: CA":'State', 'Alcohol Only':'California'})
ca_alcohol_df
```

```python
# Change unknowns to category unknowns
ca_alcohol_df.iloc[5, ca_alcohol_df.columns.get_loc('State')] = "Unknown_gender"
ca_alcohol_df.iloc[20, ca_alcohol_df.columns.get_loc('State')] = "Unknown_age"
ca_alcohol_df.iloc[28, ca_alcohol_df.columns.get_loc('State')] = "Unknown_race"
ca_alcohol_df.iloc[33, ca_alcohol_df.columns.get_loc('State')] = "Unknown_ethnicity"
ca_alcohol_df.iloc[6, ca_alcohol_df.columns.get_loc('State')] = "Total_gender"
ca_alcohol_df.iloc[21, ca_alcohol_df.columns.get_loc('State')] = "Total_age"
ca_alcohol_df.iloc[29, ca_alcohol_df.columns.get_loc('State')] = "Total_race"
ca_alcohol_df.iloc[34, ca_alcohol_df.columns.get_loc('State')] = "Total_ethnicity"
```

```python
ca_alcohol_df
```

```python
indexes_to_drop = [2, 7, 22, 30]
ca_alcohol_df = ca_alcohol_df.drop(indexes_to_drop, axis=0)
ca_alcohol_df
```
