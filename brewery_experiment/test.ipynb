```python
import requests
import json
from pprint import pprint
```

```python
from splinter import Browser
from bs4 import BeautifulSoup as soup
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
browser = Browser('chrome')
```

```python
url = "https://wisevoter.com/state-rankings/alcohol-consumption-by-state/"
browser.visit(url)
```

```python
html = browser.html
soup = soup(html, 'html.parser')
```

```python
headings = soup.find_all('th', class_="sorting")

heading_list = []

for heading in headings[1:]:
    name = heading.text
    heading_list.append(name)

heading_list
```

```python
rows = soup.find_all("tr", class_=["even", "odd"])
row_list = []

for row in rows:
    elements = row.find_all('td')
    element_list = []
    for element in elements[1:]:
        text = element.text
        element_list.append(text)
    row_list.append(element_list)

row_list 
```

```python
df = pd.DataFrame(row_list, columns=heading_list)
df
```

```python
browser.quit()
```

```python
states = df["State"]

data = []
for state in states:
    state_name = state.replace(" ","_").lower()
    url = f"https://api.openbrewerydb.org/v1/breweries?by_state={state_name}&per_page=100"
    response = requests.get(url).json()
    data.extend(response)

pprint(data[0])
```

```python
brew_locations = [x for x in data if x['latitude'] is not None]
brew_locations[0]
```

```python
save_file = open("brew_locations.json", "w")  
json.dump(brew_locations, save_file, indent = 4)  
save_file.close()  
```
